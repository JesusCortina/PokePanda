# In this notebook, I am cleaning the data

```python
import pandas as pd
import numpy as np
```

```python
#Extracting the source database from an excel file
pokedata_xl = pd.ExcelFile("pokepanda.xlsx")
gen1_7 = pokedata_xl.parse("gen17")
gen8 = pokedata_xl.parse("gen8")
```

```python
gen8
```

```python
#Uniting the two databases 
genall = pd.concat([gen1_7, gen8], ignore_index= True)
```

```python
genall
```

```python
#Filling in the regional forms
regional = genall.loc[pd.isnull(genall["gen_introduced"])]
regional_numbers = regional.index

#Filling in with numbers the NAN
for i in list(regional_numbers):
    if pd.isnull(genall.loc[i, "gen_introduced"]) == True:
        gen_intro = input("Please write the generation in which the pokemon was introduced. Information: NAME: {}, Type : {} {}".format(genall.loc[i,"pokemon"], genall.loc[i,"type1"], genall.loc[i,"type2"]))
        genall.loc[i, "gen_introduced"] = float(gen_intro)
        
genall.loc[regional_numbers]
```

```python
#Creating a region column with mapping
regionmap = {1 : "Kanto", 2 : "Johto", 3 : "Hoenn", 4 : "Sinnoh", 5 : "Unova", 6 : "Kalos", 7 : "Alola", 8 : "Galar"}
genall["region"] = genall["gen_introduced"].map(regionmap)
genall.loc[50:65]
```

```python
#Creating a column for total stats
genall["totalstats"] = genall["hp"] + genall["atk"]  + genall["def"] + genall["spatk"] + genall["spdef"] + genall["spe"]
```

```python
#Combining both types columns
genall["type2"] = genall["type2"].fillna(0)
genall["type2"] = genall["type2"].replace(0, "", regex = True)
genall["Types"] = genall[["type1", "type2"]].values.tolist()
genall
```

```python
#Combining all abilities columns
genall["ability2"] = genall["ability2"].fillna(0)
genall["ability2"] = genall["ability2"].replace(0, "", regex = True)
genall["hability"] = genall["hability"].fillna(0)
genall["hability"] = genall["hability"].replace(0, "", regex = True)
genall["Abilities"] = genall[["ability1", "ability2", "hability"]].values.tolist()
genall
```

```python
#Creating type dummies
typesiter = (set(x) for x in genall.Types)
types = sorted(set.union(*typesiter))
typedummies = pd.DataFrame(np.zeros((len(genall), len (types))), columns = types, index= genall.index)
for i, type in enumerate(genall.Types):
    for y in type:
        typedummies.loc[i, y] = 1
genall_plus = genall.join(typedummies.add_prefix("Type:"))
genall_plus = genall_plus.rename(columns = {'Type:': "Monotype"})
```

```python
#Creating region dummies
region_dumm = pd.get_dummies(genall_plus["region"])
genall_plus2 = genall_plus.join(region_dumm.add_prefix("Region:"))
genall_plus2
```

```python
#Function for creating new entries
def new_pokemon_create(ndex, pokemon, type1, gen, ability1 ,  
                hp, atk, def1, spatk, spdef, spe, type2 = "", ability2 = "", hability = "", a_bug = 1, a_dark = 1, a_dragon = 1, a_electric = 1,
                a_fairy = 1, a_fight = 1, a_fire = 1, a_flying = 1,
                a_ghost = 1, a_grass = 1, a_ground = 1, a_ice = 1,
                a_normal = 1, a_poison = 1, a_psychic = 1, a_rock = 1,
                a_steel = 1, a_water = 1):
    #function to create new pokemon. For the weaknesses, specify only those that aren't neutral using a_type. Ex. A weakness to ice goes like "a_ice = 2"
    regionmap = {1 : "Kanto", 2 : "Johto", 3 : "Hoenn", 4 : "Sinnoh", 5 : "Unova", 6 : "Kalos", 7 : "Alola", 8 : "Galar"}
    region = regionmap[gen]
    totalstats = hp + atk + def1 + spatk + spdef + spe
    if type2 == "":
        monotype = 1
    else:
        monotype = 0    
    t_list = [type1, type2]
    type_list = {"Type:Bug" : 0,
             'Type:Dark' : 0,
             'Type:Dragon' : 0,
             'Type:Electric' : 0,
             'Type:Fairy' : 0,
             'Type:Fighting' : 0,
             'Type:Fire' : 0,
             'Type:Flying' : 0,
             'Type:Ghost' : 0,
             'Type:Grass' : 0,
             'Type:Ground' : 0,
             'Type:Ice' : 0,
             'Type:Normal' : 0,
             'Type:Poison' : 0,
             'Type:Psychic' : 0,
             'Type:Rock' : 0,
             'Type:Steel' : 0,
             'Type:Water' : 0}
    for y in t_list:
        for key in list(type_list.keys()):
            if "Type:{}".format(y) == key:
                type_list[key] = 1  
    region_dict = {'Region:Alola' : 0,
                'Region:Galar' : 0,
                'Region:Hoenn' : 0,
                'Region:Johto' : 0,
                'Region:Kalos' : 0,
                'Region:Kanto' : 0,
                'Region:Sinnoh' : 0,
                'Region:Unova' : 0}
    for key in list(region_dict.keys()):
        if "Region:{}".format(region) == key:
            region_dict[key] = 1
    new_poke = {'ndex':ndex, 
                'pokemon' : pokemon, 
                'type1' : type1, 
                'type2' : type2, 
                'gen_introduced' : float(gen), 
                'ability1' : ability1, 
                'ability2' : ability2 , 
                'hability' : hability, 
                'hp' : hp, 
                'atk' : atk,
                'def' : def1, 
                'spatk' : spatk,
                'spdef' : spdef,
                'spe' : spe,
                'against_bug' : float(a_bug), 
                'against_dark' : float(a_dark),
                'against_dragon' : float(a_dragon),
                'against_electric' : float(a_electric),
                'against_fairy' : float(a_fairy), 
                'against_fight' : float(a_fight), 
                'against_fire' : float(a_fire),
                'against_flying' : float(a_flying),
                'against_ghost' : float(a_ghost), 
                'against_grass' : float(a_grass),
                'against_ground' : float(a_ground), 
                'against_ice' : float(a_ice),
                'against_normal' : float(a_normal),
                'against_poison' : float(a_poison),
                'against_psychic' : float(a_psychic), 
                'against_rock' : float(a_rock),
                'against_steel' : float(a_steel),
                'against_water' : float(a_water),
                'region' : region, 
                'totalstats' : totalstats, 
                'Types' : [type1, type2],
                'Abilities' : [ability1, ability2, hability],
                'Monotype' : monotype , 
                'Type:Bug' : type_list["Type:Bug"],
                'Type:Dark' : type_list["Type:Dark"],
                'Type:Dragon' : type_list["Type:Dragon"],
                'Type:Electric' : type_list["Type:Electric"],
                'Type:Fairy' : type_list["Type:Fairy"],
                'Type:Fighting' : type_list["Type:Fighting"],
                'Type:Fire' : type_list["Type:Fire"],
                'Type:Flying' : type_list["Type:Flying"],
                'Type:Ghost' : type_list["Type:Ghost"],
                'Type:Grass' : type_list["Type:Grass"],
                'Type:Ground' : type_list["Type:Ground"],
                'Type:Ice' : type_list["Type:Ice"],
                'Type:Normal' : type_list["Type:Normal"],
                'Type:Poison' : type_list["Type:Poison"],
                'Type:Psychic' : type_list["Type:Psychic"],
                'Type:Rock' : type_list["Type:Rock"],
                'Type:Steel' : type_list["Type:Steel"],
                'Type:Water' : type_list["Type:Water"],
                'Region:Alola' : region_dict['Region:Alola'],
                'Region:Galar' : region_dict['Region:Galar'],
                'Region:Hoenn' : region_dict['Region:Hoenn'],
                'Region:Johto' : region_dict['Region:Johto'],
                'Region:Kalos' : region_dict['Region:Kalos'],
                'Region:Kanto' : region_dict['Region:Kanto'],
                'Region:Sinnoh' : region_dict['Region:Sinnoh'],
                'Region:Unova' : region_dict['Region:Unova']}
    return new_poke

    
```

```python
#Testing the function
Calyrex = new_pokemon_create(999, "Calyrex", type1 = "Psychic", gen = 8, ability1 = "Pressure",  
                hp = 100, atk = 100, def1 = 100, spatk = 100, spdef = 100, spe = 100, type2 = "Grass", 
                a_bug = 4, a_dark = 2, a_electric = 0.5,
                a_fight = 0.5, a_fire = 2, a_flying = 2,
                a_ghost = 2, a_grass = 0.5, a_ground = 0.5, a_ice = 2,
                a_poison = 2, a_psychic = 0.5,
                a_water = 0.5)
testappend = genall_plus2.append(Calyrex, ignore_index= True)
```
